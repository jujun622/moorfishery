```python
%reload_ext autoreload
%autoreload 2

from rnn_surplus import *

c = 50
num = 50
B0 = 5000
q = 0.03
r = 3.5
m = 2
K = 10000

biomasses, efforts, catches = datasets.sample_surplus(r, m, K, B0, q, c, num, noise=False)

check = datasets.check_negative(biomasses, efforts, catches)
if check == False:
    for seed in [1,2,3,4,5]:
        print('')
        print('seed=',seed)
        net = FishRNN()
        optimizer = torch.optim.LBFGS(net.parameters(), lr=0.5)
    #     optimizer = torch.optim.SGD(net.parameters(), lr=0.01)
        best_solution = train(net, optimizer, efforts, catches, num_epochs=15, batch_size=1, init=True, random_init=True, noise=False,
              verbose=False, random_state=seed)
        sol = net.r.data.item(), net.K.data.item(), net.B0.data.item(), net.q.data.item()
        print('true =', r, m, K, B0, q, '; mse=',
              expected_mse(r, m, K, B0, q, efforts, catches, noise=False, nsample=1, retStd=True))
        print('sol =', sol, '; mse =',
              expected_mse(sol[0], m, sol[1], sol[2], sol[3], efforts, catches, noise=False, nsample=1000,
                           retStd=True))
        print('best solution = ', best_solution, '; mse =',
              expected_mse(best_solution[0], m, best_solution[1], best_solution[2], best_solution[3], efforts, catches, noise=False, nsample=1000,
                           retStd=True))
```

```python
%reload_ext autoreload
%autoreload 2

from rnn_surplus import *

c = 50
num = 50
B0 = 5000
q = 0.03
r = 3.5
m = 2
K = 10000

biomasses, efforts, catches = datasets.sample_surplus(r, m, K, B0, q, c, num, noise=False)

check = datasets.check_negative(biomasses, efforts, catches)
if check == False:
    net = FishRNN()
    optimizer = torch.optim.LBFGS(net.parameters(), lr=0.5)
    best_sol, best_mse = find_sol(model=net, optimizer=optimizer, efforts=efforts, catches=catches, num_epochs=15, batch_size=1, init=True, random_init=True, noise=False,
             verbose=False, random_state=None, trials=10)
    print(best_sol)
    print(best_mse)
```

```python
train(net, optimizer, efforts, catches, num_epochs=25, batch_size=1, init=False, random_init=False, noise=False,
          optim_closure=True, verbose=True)
sol = net.r.data.item(), net.K.data.item(), net.B0.data.item(), net.q.data.item()
print('true =', r, m, K, B0, q, '; mse=',
      expected_mse(r, m, K, B0, q, efforts, catches, noise=False, nsample=1, retStd=True))
print('sol =', sol, '; mse =',
      expected_mse(sol[0], m, sol[1], sol[2], sol[3], efforts, catches, noise=False, nsample=1000,
                   retStd=True))
```

```python
c = 50
num = 50
B0 = 3000
q = 0.05
r = 2.5
m = 2
K = 10000

biomasses, efforts, catches = datasets.sample_surplus(r, m, K, B0, q, c, num, noise=False)

check = datasets.check_negative(biomasses, efforts, catches)
if check == False:
    net1 = FishRNN()
    optimizer = torch.optim.Adadelta(net.parameters(), lr=1)
    train(net1, optimizer, efforts, catches, num_epochs=15, batch_size=1, init=True, random_init=False, noise=False,
          optim_closure=False, verbose=True)
    sol = net1.r.data.item(), net1.K.data.item(), net1.B0.data.item(), net1.q.data.item()
    print('true =', r, m, K, B0, q, '; mse=',
          expected_mse(r, m, K, B0, q, efforts, catches, noise=False, nsample=1, retStd=True))
    print('sol =', sol, '; mse =',
          expected_mse(sol[0], m, sol[1], sol[2], sol[3], efforts, catches, noise=False, nsample=1000,
                       retStd=True))
else:
    print('pass')
```
